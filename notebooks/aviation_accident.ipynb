```python
# handle suspicious operator entries
suspicious_operators = df[df['operator'].str.contains(r'op\.for', na=False)]
print(suspicious_operators[['operator']])
```

```python
import pandas as pd
import re 

def clean_operator(op):
    if pd.isna(op):
        return op
    if 'op.for' in op:
        match = re.search(r'op\.for\s+(.+)$', op)
        return match.group(1).strip() if match else op
    return op.strip()

df['operator'] = df['operator'].apply(clean_operator)
```
